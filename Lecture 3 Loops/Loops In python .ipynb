## Using for Loop with Range Function 

```python
#range(start, stop, step)
numbers = range(5,20,1)
print(list(numbers))  # Output: [0, 1, 2, 3, 4]
```

```python
squares = []

for num in range(1,100):  
    square = num * num
    squares.append(square)

print(squares)
```

# Loop control statements 

## Break 

```python
fruits = ["apple", "banana", "cherry", "kiwi", "grapefruit", 
          "orange", "watermelon", "pineapple", "mango", "grape"]

for i, fruit in enumerate(fruits):
    if fruit=="orange":
        print("Orange found ")
        print("index of Orange  is ", i)
        break
    print(fruit)
    
    
```

## Continue 

```python
fruits = ["apple", "banana", "cherry", "kiwi", "grapefruit", 
          "orange", "watermelon", "pineapple", "mango", "grape"]

for i, fruit in enumerate(fruits):
    if fruit=="orange":
        print("Not priting for  ",fruit)
        continue
        
       
    print("Index of ",fruit, "is :", i)
    
    
```

# Nested Loops 

```python
##Matrix addition 

a=[[2,3,4],
   [5,8,2]]

print("A:",a)
b=[[9,3,1],
   [7,3,2]]

print("B:", b)

result=[[0,0,0],
        [0,0,0]]

for i in range(2):
    for j in range(3):
        result[i][j]=a[i][j]+b[i][j]
print("Result is :", result)
```
